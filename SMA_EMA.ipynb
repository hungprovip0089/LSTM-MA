```python
import pandas as pd
import matplotlib.pyplot as plt
import keras
import numpy as np

stockprices = pd.read_csv('AAPL.csv')['Close'][-1::-1]

window_size = 50

SMA = stockprices.rolling(window_size).mean()
EMA = stockprices.ewm(window_size).mean()

plt.figure(figsize=[18, 13.5])

plt.plot(stockprices, color = 'blue', label = 'AAPL Stock Price')
plt.plot(SMA, color = 'red', label = 'Predicted AAPL Stock Price with SMA')
plt.plot(EMA, color = 'green', label = 'Predicted AAPL Stock Price with EMA')
plt.title('TATA Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('TATA Stock Price')

loss = keras.losses.mean_absolute_error(stockprices, SMA)
mae_SMA = np.mean(loss)

loss = keras.losses.mean_absolute_error(stockprices, EMA)
mae_EMA = np.mean(loss)

plt.text(3500,600,"MAE with SMA= " + str(mae_SMA),fontsize=22)
plt.text(3500,550,"MAE with EMA= " + str(mae_EMA),fontsize=22)


plt.legend()
plt.show()
```
